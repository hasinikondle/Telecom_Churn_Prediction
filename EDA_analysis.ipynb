```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Exploratory Data Analysis (EDA):

```python
#load Telco Customer Churn dataset
#Load the dataset and display the first few rows
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
#Check dataset shape and data types
df.info()
```

```python
#checking shape of data
df.shape
```

```python
#Identify missing values
df.isnull().sum()
```

```python
#drop unwanted columns

df = df.drop("customerID",axis=1)
```

```python
#Drop duplicate rows
df.drop_duplicates(inplace = True)
```

```python
# After removing duplicates checking shape
df.shape
```

```python
#handing total charges column

#i) checking if there are any blank values exist and replacing them with none
df['TotalCharges'] = df['TotalCharges'].replace(" ", None)

#ii)converting object dtype to numerical column
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

#Identify nulll values in total_charges
df["TotalCharges"].isnull().sum()

#iv)filling null values with mean
df["TotalCharges"] = df["TotalCharges"].fillna(round(df["TotalCharges"].mean()))

#V)checking null values removed 
df["TotalCharges"].unique()
```

```python
#Shows how many customers churned vs stayed
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

The majority of customers remain with the company. Only a smaller portion of customers churn

```python
#Churn vs Gender
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

The number of male and female customers is almost equal, and the churn counts for both genders are also very similar.

```python
#Churn vs Contract Type
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

Customers with month-to-month contracts have the highest churn rate, while customers with one-year and two-year 
contracts churn much less. The churn count is especially very low for two-year contracts.

```python
#Tenure distribution
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title("Distribution of Tenure")
plt.show()
```

```python
#Payment Method vs Churn
plt.figure(figsize=(10,5))
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.xticks(rotation=45)
plt.title("Churn by Payment Method")
plt.show()
```

```python
#reation between numerical columns
sns.heatmap(df[['tenure','MonthlyCharges','TotalCharges']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

2. Data Preprocessing:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Encode categorical variables
enc = LabelEncoder()
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = enc.fit_transform(df[col])
```

Model Training

```python
# Divide data into input and output features
X = df.drop(["Churn"],axis=1)
y = df["Churn"]
```

```python
#Split the data into training and testing sets
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Xtrain = sc.fit_transform(Xtrain)
Xtest = sc.transform(Xtest)
```

Logistic Regression

```python
#model training
LR = LogisticRegression()
LR.fit(Xtrain,ytrain)
```

```python
#Predictions
ypred = LR.predict(Xtest)
ypred
```

```python
print("Training Eff: ",LR.score(Xtrain,ytrain))
print("Training Eff: ",LR.score(Xtest,ytest))
print("R2 score: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

Decision Tree


```python
DT = DecisionTreeClassifier()
DT.fit(Xtrain,ytrain)
```

```python
#Predictions
ypred = DT.predict(Xtest)
ypred
```

```python
confusion_matrix(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
print("Training Eff: ",DT.score(Xtrain,ytrain))
print("Training Eff: ",DT.score(Xtest,ytest))
print("R2 score: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```
